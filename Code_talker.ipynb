### Generate 3D facial animation with your uploaded ausio (.wav) and the selected template and style
The duration of the audio should be around 10 seconds

```python
from IPython.display import Audio
from google.colab import files

%cd /content/CodeTalker/demo/wav/
### can upload single wave file only
uploaded = files.upload()
filename = list(uploaded.keys())[0]
print(filename)
```

```python
%cd /content/CodeTalker/

### check the uploaded audio
display(Audio(os.path.join('demo/wav/'+filename)))


### select dataset, style, and subject
from ipywidgets import interact, Dropdown, Select
vocaset_style = ['FaceTalk_170728_03272_TA', 'FaceTalk_170904_00128_TA', 'FaceTalk_170725_00137_TA', 'FaceTalk_170915_00223_TA', 'FaceTalk_170811_03274_TA', 'FaceTalk_170913_03279_TA', 'FaceTalk_170904_03276_TA', 'FaceTalk_170912_03278_TA']
# vocaset_subject = ['FaceTalk_170809_00138_TA', 'FaceTalk_170731_00024_TA']
vocaset_subject = ['FaceTalk_170728_03272_TA', 'FaceTalk_170904_00128_TA', 'FaceTalk_170725_00137_TA', 'FaceTalk_170915_00223_TA', 'FaceTalk_170811_03274_TA', 'FaceTalk_170913_03279_TA', 'FaceTalk_170904_03276_TA', 'FaceTalk_170912_03278_TA','FaceTalk_170809_00138_TA', 'FaceTalk_170731_00024_TA']
biwi_style = ['F2', 'F3', 'F4', 'M3', 'M4', 'M5']
biwi_subject = ['F1', 'F5', 'F6', 'F7', 'F8', 'M1', 'M2', 'M6','F2', 'F3', 'F4', 'M3', 'M4', 'M5']

dataset = {'VOCASET':[vocaset_style,vocaset_subject],\
           'BIWI':[biwi_style,biwi_subject]}
datasetW = Select(options = dataset.keys())
styleW = Select()
subjectW = Select()

def update_otherW_options(*args): # *args represent zero (case here) or more arguments.
    styleW.options = dataset[datasetW.value][0]
    subjectW.options = dataset[datasetW.value][1]

datasetW.observe(update_otherW_options) # Here is the trick, i.e. update cityW.options based on countryW.value.
#---
@interact(dataset = datasetW, style = styleW, subject = subjectW)
def print_city(dataset, style, subject):
    print(f'Dataset:{dataset}\nStyle:{style}\nSubject:{subject}')
```

```python
print(f'Dataset:{datasetW.value}\nStyle:{styleW.value}\nSubject:{subjectW.value}\nAudio:{filename}')
if datasetW.value == 'VOCASET':
  !sed -i '41s/.*/  condition: {styleW.value}/g' config/vocaset/demo.yaml
  !sed -i '42s/.*/  subject: {subjectW.value}/g' config/vocaset/demo.yaml
  !sed -i '43s/.*/  demo_wav_path: demo\/wav\/{filename}/g' config/vocaset/demo.yaml
  !sh scripts/demo.sh vocaset
elif datasetW.value == 'BIWI':
  !sed -i '42s/.*/  condition: {styleW.value}/g' config/BIWI/demo.yaml
  !sed -i '43s/.*/  subject: {subjectW.value}/g' config/BIWI/demo.yaml
  !sed -i '44s/.*/  demo_wav_path: demo\/wav\/{filename}/g' config/BIWI/demo.yaml
  !sh scripts/demo.sh BIWI
```
